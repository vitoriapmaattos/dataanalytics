```python
import pandas as pd
```

```python
dados = pd.read_csv("/content/sih_cnv_qiuf22354845_166_204_25.csv", encoding="ISO-8859-1",  skiprows=3, sep=";", skipfooter=12, thousands=".", decimal=",")
```

#Corrigindo o desafio:

```python
dados["2023/Dez"]
```

Média de uma coluna:

```python
dados["2023/Dez"].mean()
```

Criando um gráfico:



```python
dados.plot(x="Unidade da Federação", y="2023/Dez")
```

Estilizando o gráfico

```python
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

axis = dados.plot(x="Unidade da Federação", y="2023/Dez", kind="bar", figsize=(20,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.xticks(rotation=45)

plt.title("Valor por Unidade da Federação")

plt.show()
```
