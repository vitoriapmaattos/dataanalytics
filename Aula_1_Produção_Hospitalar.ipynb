#Analisando dados da saúde Hospitalar

Importar a biblioteca pandas:

```python
import pandas as pd
```

Ler, higienizar arquivo e inserir os dados em uma variável: (criando um data frame)

```python
dados = pd.read_csv("/content/sih_cnv_qiuf22354845_166_204_25.csv", encoding="ISO-8859-1",  skiprows=3, sep=";", skipfooter=12, thousands=".", decimal=",")
```

> encoding = Define o padrãod e códigos do arquivo

> skiprows = Exclui as linhas iniciais do arquivo

> skipfooter = Exclui as linhas finais do arquivo

> sep = Define o separados do arquivo CSV

> thousands = Define a pontuação padrão para milheiros

> decimal = Define a pontuação padrão para decimais
