```python
def descobrir_regiao(estado):
  regiao_norte = ['12 Acre', '16 Amapá', '13 Amazonas', '15 Pará', '11 Rondônia', '14 Roraima', '17 Tocantins']
  regiao_nordeste = ['21 Maranhão', '22 Piauí', '23 Ceará','24 Rio Grande do Norte', '25 Paraíba', '26 Pernambuco', '27 Alagoas','28 Sergipe', '29 Bahia']
  regiao_centro_oeste = ['50 Mato Grosso do Sul', '51 Mato Grosso', '52 Goiás', '53 Distrito Federal']
  região_sudeste = ['31 Minas Gerais', '32 Espírito Santo', '33 Rio de Janeiro', '35 São Paulo']
  regiao_sul = ['41 Paraná', '42 Santa Catarina','43 Rio Grande do Sul']

  if estado in regiao_norte:
    return 'Norte'
  elif estado in regiao_nordeste:
    return 'Nordeste'
  elif estado in regiao_centro_oeste:
    return 'Centro-Oeste'
  elif estado in região_sudeste:
    return 'Sudeste'
  elif estado in regiao_sul:
    return 'Sul'

regioes = list(map(descobrir_regiao, dados_usaveis_total.index))

dados_usaveis_total['Região'] = regioes

dados_usaveis_total.head()
```

O seaborn entenderia isso se nossas colunas fosem de datas, mas elas são colunas de texto.

Veremos como fazer essa alteração de tipo.

```python
#utilizando a função map, para mapear os dados e encontrar um novo valor pra esse dato inicial (equivalente ao procv)
from datetime import date

meses = {
    "Jan" : 1,
    "Fev" : 2,
    "Mar" : 3,
    "Abr" : 4,
    "Mai" : 5,
    "Jun" : 6,
    "Jul" : 7,
    "Ago" : 8,
    "Set" : 9,
    "Out" : 10,
    "Nov" : 11,
    "Dez" : 12
}

def para_dia(ano_mes: str):
  ano = int(ano_mes[:4])
  mes_str: str = ano_mes[5:]
  mes: int = meses[mes_str]
  return date(ano, mes, 1)
```

```python
mensal.index = mensal.index.map(para_dia)
mensal.head()
```

```python
#Plotando o mesmo gráfico após a alteração de data
plt.figure(figsize=(15,6)) #tamanho do gráfico
axis = sns.lineplot(data=mensal, x=mensal.index, y="Santa Catarina") #base 1
axis = sns.lineplot(data=mensal, x=mensal.index, y="Rio Grande do Sul") #base 2

plt.grid(linestyle="--") #linhas tracejadas
```

Transformando os dados:

Nossa tabela padrão tem um coluna com a data, várias colunas uma para cada estado, e as linhas internas são os valores gastos por esse estado nessa data.

Agora iremos transforma essa planilha de muitas colunas em uma tabela de 3 colunas

**data | estado | valor gasto**

```python
#reset_index -> Cria um novo index padrã (1, 2, 3, 4...)

mensal_aberto = mensal.reset_index().melt(id_vars=["index"], value_vars=mensal.columns)
mensal_aberto.columns = ["dia_mes_ano", "uf", "gasto"]
mensal_aberto.head()
```

```python
plt.figure(figsize=(15,6)) #tamanho do gráfico
axis = sns.lineplot(data=mensal_aberto, x="dia_mes_ano", y="gasto", hue="uf") #HUE define a cor por estado

plt.grid(linestyle="--") #linhas tracejadas
```
