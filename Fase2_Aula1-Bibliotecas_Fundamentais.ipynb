```python
#Criando uma nova função com um if

def categorizar_tomate_media(media):
  if media >= 40 and media <= 70:
    return 'Tomate médio'
  elif media < 40:
    return 'Tomate pequeno'
  else:
    return 'Tomate grande'
```

```python
#Incluindo essa função como uma coluna na base

df_csv['Categoria'] = df_csv['Average'].apply(categorizar_tomate_media)
```

```python
df_csv.groupby('Categoria').describe() #agrupando pela coluna categoria e informando os dados estatisticos
```

**Biblioteca 2: Numpay**

```python
#Concatenando várias listas

arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6], [7,8]])
arr3 = np.array([[9,10], [11,12]])

arr4 = np.concatenate((arr1, arr2, arr3), axis=1)
print(arr4)
```

```python
arr4_split = np.split(arr4, 2)
print(arr4_split) #Dividindo o array em 2 partes
```

```python
arr4_transpose = np.transpose(arr4) #Transpondo o ARR4
print(arr4_transpose)
```

```python
#Revertendo o transpose
arr4_revertido = arr4_transpose.T
arr4_revertido
```

```python
import sklearn #Importando o scikit-learn e apelidando ele de sklearn
```

```python
df_diabetes = pd.read_csv('diabetes.csv', sep=',') #Importanto csv de diabetes
```

```python
df_diabetes.head()
```

```python
#Importanto uma parte da biblioteca para fazer a separação em base de treino e de teste
from sklearn.model_selection import train_test_split
```

```python
x = df_diabetes.drop(['Class variable'], axis=1) #Separando as variáveis independentes
y = df_diabetes['Class variable'] #Separando a variável dependente - Variável Target
```

# Módulo 1 - Aula 9

**Biblioteca 4: Scikit-Learn**

```python
#Separando os dados em base de treino e teste

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
#test_size é o tamanho da amostra da base de teste
```

```python
len(x_train)
```

```python
df_diabetes.shape
```

```python
len(x_test)
```

**KNN - Primeiro modelo**

*O algoritmo opera calculando a distância entre o ponto de teste (cuja etiqueta
ou valor se deseja prever) e cada ponto de dados no conjunto de treinamento. Depois, ele seleciona os 'k' pontos mais próximos e faz uma previsão com base nas etiquetas ou valores desses vizinhos.*

```python
from sklearn.neighbors import KNeighborsClassifier #Importando o KNN
```

```python
#Parametros do modelo

knn = KNeighborsClassifier(n_neighbors=3) #n_neighbors é o número de vizinhos
```

```python
knn.fit(x_train, y_train) #Treinando o modelo
```

```python
#Testando a acuracia do modelo

accuracy = knn.score(x_test, y_test) #Acurácia do modelo
print((accuracy)*100, '%') #Exibindo a acuracia
```
